```python
#import libraries
import numpy as np
import requests
import pandas as pd
import datetime
```

```python
# Retrieve data via an API request
spacex_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/API_call_spacex_api.json'
response = requests.get(spacex_url)
print(response.status_code)  # 200 
```

```python
data = response.json()    # Convert the JSON response into a Python dictionary
df = pd.json_normalize(data)
df.head()                  
```

```python
rocket_id = df['rocket'][0]
response = requests.get(f"https://api.spacexdata.com/v4/rockets/{rocket_id}").json()

print(response)
```

```python
# Uses the rocket column to call the SpaceX API and append the rocket name to the list 
def getBoosterVersion(df):
    BoosterVersion = []
    for rocket_id in df['rocket']:
        if rocket_id:
            response = requests.get(f"https://api.spacexdata.com/v4/rockets/{rocket_id}").json()
            BoosterVersion.append(response.get('name'))
        else:
            BoosterVersion.append(None)
    return BoosterVersion


```

```python
launchpad_id = df['launchpad'][0]
response = requests.get("https://api.spacexdata.com/v4/launchpads/" + launchpad_id).json()
print(response)
```

```python
# Takes the dataset and uses the launchpad column to call the API and append the data to the list
def getLaunchSite(data):

    Longitude = []
    Latitude = []
    LaunchSite = []
    
    for x in data['launchpad']:
        if x is not None:
            response = requests.get("https://api.spacexdata.com/v4/launchpads/" + str(x)).json()
            Longitude.append(response.get('longitude'))
            Latitude.append(response.get('latitude'))
            LaunchSite.append(response.get('name'))
    return Longitude, Latitude, LaunchSite
```

```python
payload_id = df['payloads'][0][0]
response = requests.get("https://api.spacexdata.com/v4/payloads/" + payload_id).json()
print(response)
```

```python
# Takes the dataset and uses the payloads column to call the API
# and append payload mass and orbit type to the lists
def getPayloadData(data):

    PayloadMass = []
    Orbit = []
    for load in data['payloads']:
        if load is not None:
            response = requests.get("https://api.spacexdata.com/v4/payloads/" + load).json()
            PayloadMass.append(response.get('mass_kg'))
            Orbit.append(response.get('orbit'))

    return PayloadMass,Orbit
```

```python
core_id = df['cores'][0][0]['core']
response = requests.get("https://api.spacexdata.com/v4/cores/" + core_id).json()
print(response)
```

```python
# Retrieve core details from the SpaceX API using core IDs
def getCoreData(data):
    Block = []
    ReusedCount = []
    Serial = []
    Outcome = []
    Flights = []
    GridFins = []
    Reused = []
    Legs = []
    LandingPad = []
    
    for core in data['cores']:
        # Check if a core ID exists
        if core['core'] is not None:
            # Call the SpaceX API to get detailed core information
            response = requests.get(
                "https://api.spacexdata.com/v4/cores/" + core['core']
            ).json()
            
            # Append relevant fields to the lists
            Block.append(response.get('block'))
            ReusedCount.append(response.get('reuse_count'))
            Serial.append(response.get('serial'))
        else:
            # Append None if no core ID is available
            Block.append(None)
            ReusedCount.append(None)
            Serial.append(None)

# extract launch-level info directly from the launch data
    for core in data['cores']:
        Outcome.append(str(core.get('landing_success')) + ' ' + str(core.get('landing_type')))
        Flights.append(core.get('flight'))
        GridFins.append(core.get('gridfins'))
        Reused.append(core.get('reused'))
        Legs.append(core.get('legs'))
        LandingPad.append(core.get('landpad'))

 # Return all lists
    return Block, ReusedCount, Serial, Outcome, Flights, GridFins, Reused, Legs, LandingPad
```

```python
df = df[['rocket', 'payloads', 'launchpad', 'cores', 'flight_number', 'date_utc']]

# We will remove rows with multiple cores because those are falcon rockets with 2 extra rocket boosters and rows that have multiple payloads in a single rocket.
df = df[df['cores'].map(len)==1]
df = df[df['payloads'].map(len)==1]

# Since payloads and cores are lists of size 1 we will also extract the single value in the list and replace the feature.
df['cores'] = df['cores'].map(lambda x : x[0])
df['payloads'] = df['payloads'].map(lambda x : x[0])

df['date'] = pd.to_datetime(df['date_utc']).dt.date

df = df[df['date'] <= datetime.date(2020, 11, 13)]

print(df.head())
```

```python
# Booster Version
BoosterVersion = getBoosterVersion(df)

# Launch Site
Longitude, Latitude, LaunchSite = getLaunchSite(df)

# Payload Data
PayloadMass, Orbit = getPayloadData(df)

# Core Data
Block, ReusedCount, Serial, Outcome, Flights, GridFins, Reused, Legs, LandingPad = getCoreData(df)
```

```python
launch_dict = {'FlightNumber': list(df['flight_number']),
'Date': list(df['date']),
'BoosterVersion':BoosterVersion,
'PayloadMass':PayloadMass,
'Orbit':Orbit,
'LaunchSite':LaunchSite,
'Outcome':Outcome,
'Flights':Flights,
'GridFins':GridFins,
'Reused':Reused,
'Legs':Legs,
'LandingPad':LandingPad,
'Block':Block,
'ReusedCount':ReusedCount,
'Serial':Serial,
'Longitude': Longitude,
'Latitude': Latitude}
```

```python
dataframe = pd.DataFrame(launch_dict)
```

```python
data_falcon9 = dataframe[dataframe['BoosterVersion'] != 'Falcon 1'].copy()  
data_falcon9.reset_index(drop=True, inplace=True)  

# Reset FlightNumber column
data_falcon9['FlightNumber'] = list(range(1, data_falcon9.shape[0]+1))

print(data_falcon9.head())
```

```python
# Replace missing PayloadMass with mean
mean_payload = data_falcon9['PayloadMass'].mean()
data_falcon9['PayloadMass'].replace(np.nan, mean_payload, inplace=True)

# Verify missing values are gone for PayloadMass
print(data_falcon9['PayloadMass'].isnull().sum())

# LandingPad can remain None
print(data_falcon9['LandingPad'].isnull().sum())
```

```python
# Export to CSV
data_falcon9.to_csv('dataset_part_1.csv', index=False)
```
